# Clinical NLP Pipeline Demo
Demonstrates BioBERT-based text classification and NER on clinical text.

```python
from transformers import AutoTokenizer, pipeline
import pandas as pd

MODEL_NAME = 'emilyalsentzer/Bio_ClinicalBERT'
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

sample_notes = [
    'Patient presents with acute chest pain radiating to left arm, diaphoresis, and shortness of breath.',
    'Type 2 diabetes mellitus poorly controlled. HbA1c 9.2%. Started on metformin 1000mg BID.',
    'Post-op day 2 after laparoscopic appendectomy. No fever. Wound healing well.',
]

ner = pipeline('ner', model=MODEL_NAME, tokenizer=tokenizer, aggregation_strategy='simple')

for note in sample_notes:
    print(f'Note: {note[:60]}...')
    entities = ner(note)
    for ent in entities:
        print(f'  [{ent["entity_group"]}] {ent["word"]} (score: {ent["score"]:.3f})')
    print()
```
